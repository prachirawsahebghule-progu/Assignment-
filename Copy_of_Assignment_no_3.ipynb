```python
#1.Create a class BankAccount with the following:

# a) Attributes: account_holder, account_number, and balance.

#b) Methods: deposit(amount), withdraw(amount), and get_balance().

#c) withdraw() should print a message and do nothing if the amount exceeds the current balance.

#d) A ___str__ method that returns a readable summary of the account.

#Create two BankaAccount object and perform a few deposits and withdrawals on each.
```

```python
class BankAccount:
    def __init__(self, account_holder, account_number, balance):
        self.account_holder = account_holder
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited ${amount}")

    def withdraw(self, amount):
        if amount > self.balance:
           print("Insufficient Balance!")
        else:
           self.balance -= amount
           print(f"Withdrawn ${amount}")

    def get_balance(self):
        return self.balance

    def __str__(self):
        return f"Account Holder: {self.account_holder}\nAccount Number: {self.account_number}\nBalance: ${self.balance}"


# Creating two accounts
acc1 = BankAccount("Prachi", "2001", 7000)
acc2 = BankAccount("divya", "2002",  9000)

# Transaction on account 1
acc1.deposit(2000)
acc2.withdraw(12000)

# Display account details
print("\nAccount 1 Details:")
print(acc1)

print("\nAccount 2 Details:")
print(acc2)
```

```python
#2. Create a class Student with attributes name, roll_no, and marks (a list of 5 subject marks). Add method:

#a) calculate_percentage() - returns the average of all marks.

#b) get_grade() - return 'A' for >= 80, 'B' for >= 60, 'C' for >= 40, and 'Fail' otherwise.

#c) display() - prints all details including percentage and grade.

#Create a list of 3 Student object and call display() on each.
```

```python
class Student:
    def __init__(self, name, roll_no, marks):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks

    def calculate_percentage(self):
        return sum(self.marks) / len(self.marks)

    def get_grade(self):
        percentage = self.calculate_percentage()

        if percentage >= 80:
            return 'A'
        elif percentage >= 60:
            return 'B'
        elif percentage >= 40:
            return 'C'
        else:
            return 'Fail'

    def display(self):
        print("Name:", self.name)
        print("Roll No:", self.roll_no)
        print("Marks:", self.marks)
        print("Percentage:", self.calculate_percentage())
        print("Grade:", self.get_grade())
        print("-" * 30)

students = [
    Student("Prachi", 1, [85, 90, 88, 92, 80]),
    Student("Divya", 2, [70, 65, 75, 68, 72]),
    Student("Priyanka", 3, [40, 45, 50, 35, 42])
]

for s in students:
    s.display()
```

```python
#Section-B- Encapsulation
#3.Create a class Employee with private attributes __name, __emp_id, and __salary. Add:

#a) Getter and setter methods for all three attributes.

#b) The salary setter should raise a ValueError if the value is below 10000.

#c) A method give_raise(percent) that increases salary by the given percentage.

#Demonstrate that the private attributes cannot be accessed directly, and that all access goes through the methods.
```

```python
class Employee:
    def __init__(self, name, emp_id, salary):
        self.__name = name
        self.__emp_id = emp_id
        self.__salary = salary

    def get_name(self):
        return self.__name

    def set_name(self, name):
        self.__name = name

    def get_emp_id(self):
        return self.__emp_id

    def set_emp_id(self, emp_id):
        self.__emp_id = emp_id

    def get_salary(self):
        return self.__salary

    def set_salary(self, salary):
        if salary < 10000:
            raise ValueError("Salary cannot be below 10000")
        self.__salary = salary

    def give_raise(self, percent):
        self.__salary += self.__salary * percent / 100

emp = Employee("Prachi", 777, 70000)

print("Name:", emp.get_name())
print("Salary:", emp.get_salary())

emp.give_raise(10)

print("Salary after raise:", emp.get_salary())
```

```python
#4.Design a class SecureVault to simulate a password-protected storage system.

#a) The class should have a private attribute __pin and a private dictionary __data.

#b) Add a method add_entry(pin, key, value) that stores the value only if the correct PIN is provided.

#c) Add a method get_entry(pin, key) that returns the value only if the correct PIN is provided.

#d) If the PIN is incorrect, print "Access Denied".
```

```python
class SecureVault:
    def __init__(self, pin):
        self.__pin = pin
        self.__data = {}

    def add_entry(self, pin, key, value):
        if pin == self.__pin:
            self.__data[key] = value
            print("Entry Added")
        else:
            print("Access Denied")

    def get_entry(self, pin, key):
        if pin == self.__pin:
            return self.__data.get(key)
        else:
            print("Access Denied")

vault = SecureVault("1107")

vault.add_entry("1107", "Password", "Megha223")
vault.add_entry("2306", "Email", "Prachi@gmail.com")

print(vault.get_entry("1107", "Password"))
print(vault.get_entry("2306", "Password"))
```

```python
# Section- C- Inheritence & method overiding
#5.Create a base class Shape with methods area() and describe().

#a) Create three subclasses: Circle, Rectangle, and Triangle.

#b) Override the area() method in each subclass to calculate the appropriate area.

#c) Override the describe() method to print a message specific to each shape.

#d) Create one object of each subclass and demonstrate polymorphism by storing them in a list and calling describe() and area() in a loop.

#Use inheritance and method overriding appropriately.
```

```python
import math

class Shape:
    def area(self):
        return 0

    def describe(self):
        print("I am a shape")

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius * self.radius

    def describe(self):
        print("I am a Circle")

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def describe(self):
        print("I am a Rectangle")

class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height

    def describe(self):
        print("I am a Triangle")

c = Circle(6)
r = Rectangle(5, 4)
t = Triangle(11, 9)

c.describe()
print("Area =", c.area())

r.describe()
print("Area =", r.area())

t.describe()
print("Area =", t.area())
```

```python
#6. Create a class Vehicle with attributes make, model, and year and a method start_engine() that prints 'Engine started'. Create two subclasses:**

#**a) Car(make, model, year, num_doors)** — overrides `start_engine()` to include the car's name in the message.

#**b) ElectricCar(make, model, year, battery_capacity)** — overrides `start_engine()` to print **'Silent start — battery at X kWh'**.

#**Use `super()` in both subclasses to call the parent `__init__()`. Create objects and test all methods.**
```

```python
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def start_engine(self):
        print("Engine started.")

class Car(Vehicle):
    def __init__(self, make, model, year, num_doors):
        super().__init__(make, model, year)
        self.num_doors = num_doors

    def start_engine(self):
        print(f"{self.make} car engine started.")

class ElectricCar(Vehicle):
    def __init__(self, make, model, year, battery_capacity):
        super().__init__(make, model, year)
        self.battery_capacity = battery_capacity

    def start_engine(self):
        print(f"Silent start — battery at {self.battery_capacity} kWh.")

car = Car("Toyota", "Corolla", 2020, 4)
ev = ElectricCar("Tesla", "Model 3", 2023, 77)

car.start_engine()
ev.start_engine()
```

```python
#Section D — Polymorphism
#7.Create a base class Animal with a method sound() that prints “Some sound”.
#Create subclasses Dog, Cat, and Cow, each overriding sound() with their respective sounds.
    #Write a function make_all_sounds(animals) that takes a list of Animal objects and calls sound() on each.
#Demonstrate that the correct method is called regardless of the object type
```

```python
class Animal:
    def sound(self):
        print("Some sound")

class Dog(Animal):
    def sound(self):
        print("Bark")

class Cat(Animal):
    def sound(self):
        print("Meow")

class Cow(Animal):
    def sound(self):
        print("Moo")

def make_all_sounds(animals):
    for animal in animals:
        animal.sound()

animals = [Dog(), Cat(), Cow()]
make_all_sounds(animals)
```

```python
#Section F — Menu-Driven App with OOP
#10. Build a menu-driven Student Record Manager using OOP and file handling.
    #The app should have a class StudentManager with the following menu options:
#1.Add Student — takes name, roll number, and marks for 3 subjects.
#2.View All — displays all stored students with their average marks.
#3.Search Student — finds a student by roll number.
#4.Delete Student — removes a student by roll number.
#5.Save to File — writes all records to students_data.txt.
#6.Load from File — reads records back from students_data.txt on startup.
#7.Exit
```

```python
class StudentManager:
    def __init__(self):
        self.students = []

    def add_student(self):
        name = input("Enter Name: ")
        roll = input("Enter Roll No: ")

        m1 = float(input("Subject 1 Marks: "))
        m2 = float(input("Subject 2 Marks: "))
        m3 = float(input("Subject 3 Marks: "))

        avg = (m1 + m2 + m3) / 3

        self.students.append({
            "name": name,
            "roll": roll,
            "average": avg
        })

    def view_students(self):
        for s in self.students:
            print(s)

    def search_student(self):
        roll = input("Enter Roll No: ")
        for s in self.students:
            if s["roll"] == roll:
                print(s)
                return
        print("Student not found")

    def delete_student(self):
        roll = input("Enter Roll No: ")
        for s in self.students:
            if s["roll"] == roll:
                self.students.remove(s)
                print("Deleted")
                return
        print("Student not found")

    def save_to_file(self):
        with open("students_data.txt", "w") as f:
            for s in self.students:
                f.write(f"{s['name']},{s['roll']},{s['average']}\n")

    def load_from_file(self):
        try:
            with open("students_data.txt", "r") as f:
                for line in f:
                    name, roll, avg = line.strip().split(",")
                    self.students.append({
                        "name": name,
                        "roll": roll,
                        "average": float(avg)
                    })
        except FileNotFoundError:
            pass

manager = StudentManager()
manager.load_from_file()

while True:
    print("\n1.Add Student")
    print("2.View Students")
    print("3.Search Student")
    print("4.Delete Student")
    print("5.Save to File")
    print("6.Exit")

    choice = input("Enter Choice: ")

    if choice == "1":
        manager.add_student()
    elif choice == "2":
        manager.view_students()
    elif choice == "3":
        manager.search_student()
    elif choice == "4":
        manager.delete_student()
    elif choice == "5":
        manager.save_to_file()
    elif choice == "6":
        manager.save_to_file()
        print("Exiting...")
        break
    else:
        print("Invalid Choice")
```
