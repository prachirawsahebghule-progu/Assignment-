```python
#1.write a function analysis _sentence(sentence) that takes a sentence and returns a dictionary with word_count,
#char_count(excluding spaces). unique_words(case insensitive), and mist_frequent_word text it with
"The quick brown fox jumps over the lazy dog"
```

```python
def analyze_sentence(sentence):
    words = sentence.lower().split()

    word_count = len(words)
    char_count = len(sentence.replace(" "," "))
    unique_words = len(set(words))

    freq = {}
    for word in words:
        freq[word] = freq.get(word, 0) + 1

    most_frequent_word = max(freq, key=freq.get)

    return {
        "word_count": word_count,
        "char_count": char_count,
        "unique_words": unique_words,
        "most_frequent_word": most_frequent_word
    }
sentence = "The quick brown fox jumps over the lazy dog the fox"
print(analyze_sentence(sentence))
```

```python
#3.write a function format_report(data) that takes a list of tuples(name,score,grade)and prints a formatted table where name is left-aligned(width 15)
#score is right -aligned (width(width 6),and grade is centred(width 8).Use this data.
data=[("Alice",92,"A"),("Bob",78,"B+"),("Charlie",55,"C"),("Diana",100,"A+")]
```

```python
def format_report(data):
    print(f"{'Name':<15}{'Score':>6}{'Grade':^8}")
    print("-" * 29)

    for name, score, grade in data:
        print(f"{name:<15}{score:>6}{grade:^8}")


data = [
    ("Alice", 92, "A"),
    ("Bob", 78, "B+"),
    ("Charlie", 55, "C"),
    ("Diana", 100, "A+")
]

format_report(data)
```

```python
#7. Write a function inventory_manager(records) that takes a list of tuples(product, category, quantity, price) and returns a dictionary grouped bycategory.
#Each category should have total_items, total_value, and a listof product names.
records = [("Laptop", "Electronic", 10, 55000),("Mouse", "Electronic", 50, 999),("Desk", "Furniture", 8, 12000),("Chair", "Furniture", 15, 4500),
   ("Pen", "Stationary", 100, 20)]
```

```python
def inventory_manager(records):
    result = {}

    for product, category, qty, price in records:
        if category not in result:
            result[category] = {
                "total_items": 0,
                "total_value": 0,
                "products": []
            }

        result[category]["total_items"] += qty
        result[category]["total_value"] += qty * price
        result[category]["products"].append(product)

    return result

records = [
    ("Laptop", "Electronics", 10, 55000),
    ("Mouse", "Electronics", 50, 999),
    ("Desk", "Furniture", 8, 12000),
    ("Chair", "Furniture", 15,4500),
    ("Pen", "Stationary", 100, 20)
]

print(inventory_manager(records))
```

```python
import math

dataset = [12, 45, 7, 23, 56, 89, 34, 67, 11, 90]

def mean(data):
    return sum(data) / len(data)

def variance(data):
    m = mean(data)
    return sum((x - m) ** 2 for x in data ) / len(data)

def normalize(data):
    m = mean(data)
    std = math.sqrt(variance(data))
    return [(x - m) / std for x in data]

print("Mean =", mean(dataset))
print("Variance =", variance(dataset))
print("Normalized =", normalize(dataset))


std_dev = math.sqrt(variance(dataset))
print("Standard Deviation =", std_dev)
```

```python
#a)
def make_multiplier(n):
    def multiplier(x):
        return x * n
    return multiplier

triple = make_multiplier(3)
quintuple = make_multiplier(5)

print(triple(10))
print(quintuple(10))
```

```python
#b)
import time

def timer(function):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = function(*args, **kwargs)
        end = time.time()

        print("Execution Time;", end - start)
        return result
    return wrapper

@timer
def slow_sum(n):
    return sum(i * i for i in range(1, n + 1))

print(slow_sum(100000))
```

```python
#c)
import math

def validate_positive(function):
    def wrapper(x):
        if x <= 0:
            raise ValueError("Argument must be positive")
        return function(x)
    return wrapper

@validate_positive
def safe_sqrt(x):
    return math.sqrt(x)

print(safe_sqrt(25))
```
