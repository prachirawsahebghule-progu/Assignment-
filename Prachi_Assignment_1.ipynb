```python
#1. Take two numbers as input from the user print their sum, difference, product, and quotient -each on a separate line.
#Handle division by zero with an appropriate message
```

```python
num1 = float(input("Enter first number:"))
num2 = float(input("Enter second number:"))

print("Sum =", num1 + num2)
print("Difference =", num1 - num2)
print("Product =", num1 * num2)

if num2 != 0:
    print("Quotient =", num1 / num2)
else:
    print("Division by zero is not allowed.")
```

```python
#2. Accept a temperature in celsius from the user, convert it to Fahrenheit and kelvin,and print with three values with labels
```

```python
celsius = float(input("Enter temperature in Celsius:"))
fahrenheit = (celsius * 9/5) + 32
kelvin = celsius + 273.15

print("Celsius:", celsius)
print("Fahrenheit:", fahrenheit)
print("Kelvin:", kelvin)
```

```python
#5. Take a year as input and print whether it is a leap year or not.
```

```python
year = int(input("Enter year: "))

if (year % 400 == 0) or ( year % 4 == 0 and year % 100 != 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```

```python
#6. Build a simple calculator-take two numbers and an operator (+,-,*,/)as input, perform the operation,and print the result
#print an error message for invalid operators or division by zero
```

```python
num1 = float(input("Enter first number:"))
num2 = float (input("Enter second number:"))
op = input("Enter operator (+,-,*,/):")

if op == "+":
    print("Result =",num1 + num2)
elif op == "-":
    print("Result =",num1 - num2)
elif op == "*":
    print("Result =",num1 * num2)
elif op == "/":
    if num2 != 0:
        print("Result=",num1 / num2)
    else:
        print("Division by zero not allowed.")
else:
    print("Invalid operator.")
```

```python
#8. Take a positive integer n as input and print numbers from 2 to n.
```

```python
n = int(input("Enter n:"))

for num in range (2, n + 1):
    prime = True

    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            prime = False
            break
    if prime:
       print(num, end=" ")
```

```python
#9. Keep accepting numbers from the user until they enter 0.them print the total count of numbers entered, their sum,and their average.
#Do not include 0 in the calculations.
```

```python
count = 0
total = 0

while True:
    num = int(input("Enter number (0 to stop): "))

    if num == 0:
        break

    count += 1
    total += num

if count > 0:
    average = total / count
else:
     average = 0

print("Count =", count)
print("Sum =", total)
print("Average=", average)
```

```python
#10. Take n as input and print a triangle pattern of n rows made of asterisks, where row i has(2i-1)stars centred appropriately.
```

```python
n = int(input("Enter number of row: "))

for i in range (1, n + 1):
    spaces = " " * (n - i)
    stars = "*" * (2 * i - 1)
    print(spaces + stars)
```
